```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python

PKurl="https://scrapeme.live/shop/"

PKurl_res=requests.get(PKurl)
PKurl_res
```

```python
PK=BeautifulSoup(PKurl_res.text,'html.parser') #parse the html
PK
```

```python
PK.title
```

```python
# Page links >> 48 pages
all_pages_links = []
current_link = PKurl

while True:
    all_pages_links.append(current_link)

    request2 = requests.get(current_link, verify=False)
    soup2 = BeautifulSoup(request2.text, "html.parser")

    next_btn = soup2.select_one("a.next.page-numbers") # the next button
    if next_btn and next_btn.has_attr("href"):
        current_link = next_btn["href"]
    else:
        break
```

```python
product_links = []

for url in all_pages_links:
  # Request & Soup each page a
    request_page = requests.get(url,verify=False)
    soup_page = BeautifulSoup(request_page.text, "html.parser")


    products_list = soup_page.find("ul", class_="products")
    products = products_list.find_all("li", class_="product")

    for product in products:
        a = product.find("a", class_="woocommerce-LoopProduct-link")
        if a and a.has_attr("href"):
          link = a["href"]
          product_links.append(link)
```

```python
product_links
```

```python
names = []
prices = []
links = []
descriptions = []
availabilities = []

for link in product_links:
    request_product = requests.get(link,verify=False)
    prod_soup = BeautifulSoup(request_product.text, "html.parser")

    # الاسم
    name_tag = prod_soup.find("h2", class_="woocommerce-loop-product__title")
    name = name_tag.get_text(strip=True) if name_tag else ""

    # السعر
    p = prod_soup.find("p", class_= "price")
    price = p.get_text(" ").split()[1] if p else None # took the price without £

    # رابط صفحة المنتج
    link = prod_soup.find("a")["href"]


    # الوصف
    desc_tag = prod_soup.find("div", class_="woocommerce-product-details__short-description")
    description = desc_tag.get_text(" ", strip=True) if desc_tag else ""

    # التوفّر (In stock / Out of stock)
    avail_tag = prod_soup.find("p", class_="stock")
    availability = avail_tag.get_text(strip=True) if avail_tag else "Unknown"
     # تخزين
    names.append(name)
    prices.append(price)
    links.append(link)
    descriptions.append(description)
    availabilities.append(availability)

```

```python
# data frame
data = {
    "Name": names,
    "Price": prices,
    "Link": links,
    "Description": descriptions,
    "Availability": availabilities
}

df = pd.DataFrame(data)
df
```

```python
df.to_csv('scraped_data.csv', index=False)
```
